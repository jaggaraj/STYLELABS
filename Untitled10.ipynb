```python
dt['sno']=dt['sno'].interpolate('linear')
dt['sno']=dt['sno'].astype(int)
```

```python
dt['name']=dt['name'].astype(str)
```

```python
dt['name']=dt['name'].apply(lambda x: re.sub(r'[^a-zA-Z0-9/s]','',x))
```
