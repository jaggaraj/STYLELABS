```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.svm import SVC
```

```python
dt=pd.read_csv("C:\\Users\\Dell\\Desktop\\New Microsoft Excel Worksheet.csv")
```

```python
dt.head()
```

```python
mas=StandardScaler()
```

```python
dt.iloc[:,0:10]=mas.fit_transform(dt.iloc[:,0:10])
```

```python
dt.iloc[:,0:10]
```

```python
x_train,x_test,y_train,y_test=train_test_split(dt.iloc[:,0:10],dt.iloc[:,10:11],test_size=0.2,random_state=42)
```

```python
rfc=RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
print(classification_report(y_pred,y_test))
```
